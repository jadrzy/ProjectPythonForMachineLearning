```python
age_mapping = { '18-24': 0,'25-34': 1,'35-44': 2,'45-54': 3,'55-64': 4, '65+': 5 }
education_mapping = {'Doctorate degree': 0, 'Left school at 16 years': 1, 'Left school at 17 years': 2, 'Left school at 18 years': 3, 'Left school before 16 years': 4, 'Masters degree': 5, 'Professional certificate/ diploma': 6, 'Some college or university, no certificate or degree': 7, 'University degree': 8}
gender_map = { 'Female': 0, 'Male': 1 }

model_data['Age'] = analize_data['Age'].map(age_mapping)
model_data['Education'] = analize_data['Education'].map(education_mapping)
model_data['Gender'] = analize_data['Gender'].map(gender_map)

model_data = pd.get_dummies(model_data, columns=['Country', 'Ethnicity'], drop_first=False)

bool_columns = model_data.select_dtypes(include=['bool']).columns
bool_columns = bool_columns.unique()
model_data[bool_columns] = model_data[bool_columns].astype(int)

for column in model_data.columns:
    print(f"Unique of: '{column}':")
    print(sorted(model_data[column].unique()))
    print("\n")

print(model_data.head())
model_data.to_csv('/content/drive/MyDrive/Materiały ze studiów/PMLDS/model_data.csv', index=False)

```
